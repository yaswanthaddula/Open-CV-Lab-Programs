```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

if img is None:
    print("Image not loaded. Check path carefully.")
else:
    print("Original Image:")
    cv2_imshow(img)

    # Bigger Image
    bigger = cv2.resize(img, None, fx=2, fy=2)
    print("Bigger Image:")
    cv2_imshow(bigger)

    # Smaller Image
    smaller = cv2.resize(img, None, fx=0.5, fy=0.5)
    print("Smaller Image:")
    cv2_imshow(smaller)
```

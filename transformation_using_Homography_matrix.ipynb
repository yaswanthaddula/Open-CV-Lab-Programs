```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("//content/Screenshot 2026-02-26 140552.png")

if img is None:
    print("Image not loaded")
else:
    rows, cols = img.shape[:2]

    # Source points (original corners)
    src_pts = np.float32([[0, 0],[cols-1, 0],[0, rows-1],[cols-1, rows-1]])

    # Destination points (intentionally distorted)
    dst_pts = np.float32([[50, 50],[cols-100, 20],[100, rows-50],[cols-50, rows-100]])

    H, status = cv2.findHomography(src_pts, dst_pts)
    result = cv2.warpPerspective(img, H, (cols, rows))

    print("Original Image")
    cv2_imshow(img)

    print("Homography Transformed Image")
    cv2_imshow(result)
```

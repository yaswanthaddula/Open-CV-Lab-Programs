```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

if img is None:
    print("Image not loaded")
else:
    print("Original Image:")
    cv2_imshow(img)

    # Rotate Clockwise (90 degree)
    clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    print("Clockwise Rotation:")
    cv2_imshow(clockwise)


    # Rotate Counter Clockwise (90 degree)
    anticlockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    print("Counter Clockwise Rotation:")
    cv2_imshow(anticlockwise)
```
